```python
from gensim.models import FastText
```

```python
sentences = [
    ['this', 'is', 'a', 'sample', 'sentence'],
    ['fasttext', 'embeddings', 'are', 'powerful'],
    ['another', 'example', 'sentence', 'for', 'fasttext'],
    ['gensim', 'library', 'makes', 'it', 'easy'],
    ['word', 'vectors', 'capture', 'semantic', 'meaning']
]
```

```python

# Train a FastText model
model = FastText(sentences, vector_size=256, window=5, min_count=1, workers=4, sg=1)

# Save the model for later use
model.save("fasttext.model")
```

```python
model.wv['fasttext']
```

```python
model.wv.most_similar('word')
```
